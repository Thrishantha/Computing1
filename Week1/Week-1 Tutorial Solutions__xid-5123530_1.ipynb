### Integer and floating values 

### Q1
Evaluate $y = 2x^3 + 3x + 6$ when $x = 3$ and display the answer.

```python
x = 3 # The known variable x should be defined first
# So that the value of x is in the memory to be used in an equation

# Then when you write an equation using x, the value assigned to it
# will be used in the equation

y = 2*x**3 + 3*x + 6
print("The value of function y is {} when x is {}".format(y,x))
```

### Q2
Evaluate $y = x^3 + 3x + 8$ when $x = 3.124$ and display the answer upto 2 decimal place accuracy.

```python
x = 3.124
y = x**3 + 3*x + 8
print("The value of the function y is {:.2f} when x is {}".format(y,x))
```

$a = 2$

$b = 3$

3. Check if x is greater than y, x is equal to y, or x is less than or equal to y.

```python
a = 2
b = 3
c = ((a - b**2)**0.5)
x = c.imag
d = ((a - 4*b)**0.5)
y = d.imag

e = x > y
f = x == y
g = x <= y
print("x: {:.2f}, y: {:.2f}".format(x,y))
print("e: {}, f: {}, g: {}".format(e,f,g))
```

### Q3
Ask the user to input the first name, then the second name. 
1. Check if letter "a" is in the first name. Assign the result to boolean variable $x$.
2. Check if the letter "a" is in the last name. Assign the result to boolean variable $y$. 
3. Generate a boolean answer ("True" or "False") to the following statements: "The user has letter "a" both in the first name and the last name", "The user has letter "a" only in the first name but not in the last name".

```python
first_name = input("Enter your first name: ")
second_name = input("Enter your second name: ")

x = 'a' in first_name
y = 'a' in second_name

a = x and y
b = x and not y

print("You have letter 'a' in both the first name and the second name? {}".format(a))
print("You have letter 'a' in the first name but not in the second name? {}".format(b))
```

### Q1
Create a dictionary to store following information about three students in your tutorial group:
1. Name
2. Estimated height
3. Gender
4. Age

Use the dictionary to show how you would display the name and estimated height of two of the students.

```python
students = {
    'Allen': {'height':170,'gender':'male','age':23},
    'Liang': {'height':165,'gender':'male','age':26},
    'Eli': {'height':172,'gender':'female','age':24}
}

print("Allen's height is {} cm".format(students['Allen']['height']))
print("Eli's age is {} years".format(students['Eli']['age']))
```

### Q4
Write a code to to ask a user to enter the amount of money in the bank account. Then ask the user to enter the current interest rate. Display the total balance after 10 years.


```python
money = float(input('Please enter how much money you have in the bank in (£): '))
rate = float(input('Please enter the interest rate as a percentage value (%): '))

# Notice that the input() function is inside the float() function. This will
# take the result from the input function (a string that th user enters) and 
# pass that into the float() function which will try to convert it into a float
# type (a number that has a decimal component).

balance = money*(1 + rate/100)**10
print('Your will have £{:.2f} at the end of 10 years'.format(balance))

# Notice how the balance was specially formatted when it was inserted 
# dynamically into the string.
```
