```python
import random as r

Tables = ['study table', 'soldering table', 'Kitchen table', \
          'Wooden table', 'plastic table']
N = len(Tables)
print("There are %d tables in your list" %(N))
n = r.randint(0, N-1)
print("Table %d is %s" %(n+1,Tables[n]))
```

```python
from sympy.solvers import solve #Here we first import Python libraries 
#that have already done some work to simplify your coding
from sympy import Symbol
x = Symbol('x')
y = solve(x**3 - 12*x**2 + 44*x - 48, x)
print(y)
f = x**3 - 12*x**2 + 44*x - 48
s1 = f.subs([(x,y[0])])
print(s1)
s2 = f.subs([(x,y[1])])
print(s2)
s3 = f.subs([(x,y[2])])
print(s3)
```

```python
y = (2.0 + 3.0j)
r = y.real
im = y.imag
print("The real part of y is %1.2f and the imaginary part is %1.2f" %(r,im))
```

```python
from sympy import *

x,z,g = symbols('x z g')
g = (x**2+2*x*z + x)/x
print(g)
gsimp = simplify(g)
print(gsimp)
sol = gsimp.subs([(x,1),(z,2)])
print(sol)
```

```python
import matplotlib.pyplot as plt #Do this when you want to plot curves
import numpy as np #package for scientific computing with Python
from math import sin, cos, log

x = np.arange(1,100,2) #This is how we can initiate an array

def f(x): #You can write a function like this. 
    #A function takes input variables and gives out the result
    return sin(x)

f = np.vectorize(f)  # This allows you to pass an array of numbers 
#as inputs and get an array of data as outputs

y = f(x)  # if x is your Numpy array

plt.plot(x,y,':o')
plt.show()
```
