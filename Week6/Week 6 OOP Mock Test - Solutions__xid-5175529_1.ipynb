    `[[50,0.2],[20,0.5]]`.

* Add a method to compute the weighted average mark of the student. N.B: this method should be called each time a mark is added because we want the weighted average to always be up to date!

## Question-3: Hip Hop Dance Club

You're going to write an object oriented program that helps keep track of the members' information in a hip hop dance club at Imperial. The club has regular members who can take regular dance classes organised by the club. The club also has 2 competition teams (novice and advanced) who represent the school to compete in national competitions. 

There are specific requirements for each category (regular members, novice team members, and advanced team members) to be qualified for their category. Your program has to 

1. obtain and record information from members, 

2. check if they have met those specific requirements.

### Tasks

- Collect the following information from all members:
    - Name
    - Gender
    - CID number
    - Email
    - Check if the person has purchased membership. If no, tell the person that he/she needs to purchase a membership on https://www.imperialcollegeunion.org/ to take classes at a discounted price.


- Collect the following information from a Novice in addition to what is obtained from any member:
    - Number of years hip hop dance experience the person has (cannot be negative number)
    - Preferred style (Old school, Urban, Popping, Locking, Breaking, others)
    - Check if he/she is a member. If not, tell the person that he/she has to be a club member to be able to join the team
    - Check his/her availability for training: 
            - Monday and Wednesday: 7-9pm
            - Saturday: 12-3pm
    - If the person cannot commit to these training times, tell the person that unfortunately he/she cannot join the team


- Collect the following information from a Advanced member in addition to what is obtained from any member:
    - Number of years of hip hop dance experience the person has (cannot be negative number). 
    - Preferred style (Old school, Urban, Popping, Locking, Breaking, others)
    - Check if he/she is a member. If not, tell the person that he/she has to be a club member to be able to join the team
    - Check his/her availability for training: 
            - Tuesday and Thursday: 7-9pm
            - Saturday: 3-6pm
    - If the person cannot commit to both of these training times, tell the person that unfortunately he/she cannot join the team


Hints:
- Create a class to record the overall club information, and a subclass for members. The club should be able to check if a member would like to join a team and check the requirements accordingly by prompting the user.
- Use conditions for checking inputs.
- You should be able to ask and record information for regular members, novice team members and advanced team members separately.
- You should be able to print information about the club and its members at the end.


```python
class Club:
    '''This stores information about the club, and its members'''
    
    def __init__(self, name):
        self.name = name
        self.members = []
        self.novice_team = []
        self.advanced_team = []
        
    def print_info(self):
        '''Used to print out current club members.'''
        
        print("\n\n-- The {} club --".format(self.name))
        print("\n    All members:")
        for i, m in enumerate(self.members):
            print("        {}. {}".format(i+1, m.name))

        print("\n    The Novice team:")
        for i, m in enumerate(self.novice_team):
            print("        {}. {}".format(i+1, m.name))

        print("\n    The Advanced team:")
        for i, m in enumerate(self.advanced_team):
            print("        {}. {}".format(i+1, m.name))
            
        print("\n\n")
        
    def new_member(self):
        '''Method used to have new person apply for membership.'''
        
        new_mem = Member()
        if new_mem.has_paid():
            self.members.append(new_mem)
            return new_mem
            
    def join_novice_team(self, member):
        '''Used to have member apply to join the Novice team.'''
        
        if type(member) != Member or member not in self.members:
            print("Only paid members can join the team.")  
            return
            
        Club._joining_team_questions(member)
        joined = Club._check_availability("Our training times are every Monday and Wednesday 7-9pm, and every Saturday 12-3pm. Are you available for training at these times? ")
        if joined:
            self.novice_team.append(member)
            
    
    def join_advanced_team(self, member):
        '''Used to have member apply to join the Advanced team.'''
        
        if type(member) != Member or member not in self.members:
            print("Only paid members can join the team.")  
            return
            
        Club._joining_team_questions(member)
        joined = Club._check_availability("Our training times are every Monday and Wednesday 7-9pm, and every Saturday 3-6pm. Are you available for training at these times? ")
        if joined:
            self.advanced_team.append(member)
            
    
    # We use an underscore at the beginning of the method name to indicate
    # that this method should only be used by other methods of this class
    def _joining_team_questions(member_obj):
        '''Internal method that asks for extended information about the member needed for joining a team.'''
        
        member_obj.experience = float(input("How many years have you been dancing hip pop? "))
        member_obj.genre = input("What sub-genre of hip hop do you prefer? (Old school, Urban, Popping, Locking, Breaking, others) ")

    def _check_availability(message):
        '''Internal method for prompting to check the availability of the member.'''
        
        print(message)
        avail = input("yes/no: ")
        if avail == "yes":
            print("Great. You have joined the team.")
            return True
        else:
            print("Unfortunately, if you cannot commit to these times, you can't join the team.")
            return False       


class Member:
    '''This class stores the information of a member of the club by prompting the user.'''
    
    def __init__(self):
        self.first_name = input("What is your first name? ")
        self.last_name = input("What is your last name? ")
        self.name = str(self.first_name +" "+ self.last_name)
        self.gender = input("Your gender please? (m/f) ")
        self.CID = input("What is your CID number? ")
        self.email = input("What is your college email? ")
        
        
    def has_paid(self):
        '''Asks the member whether they have paid for membership or not. 
        Returns a boolean based on the answer.'''
        
        check = input("Have you purchased our club membership? (yes/no) ")
        if check == "yes":
            print("Welcome to the club! Your input has been recorded.")
            return True
        elif check == "no":
            print("I'm sorry, you are not a member of our club yet. You can purchase your membership here https://www.imperialcollegeunion.org/ to enjoy discounted price for our classes.")
            return False
        else:
            raise ValueError("You must answer yes/no")
            

# Create the club
club = Club("IC Hip Hop")

# Add some members to the club
person1 = club.new_member()
club.join_novice_team(person1)

person2 = club.new_member()
club.join_advanced_team(person2)

# Print some information about the club we have collected
club.print_info()

# We check the documentation of our classes
print(help(Club))
print(help(Member))
```
