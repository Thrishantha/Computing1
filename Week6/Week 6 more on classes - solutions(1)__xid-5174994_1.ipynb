## Exercise 6.1

Create a class `Shape` that has the attributes `name` and `colour`, as well as a method `getInfo(self)` that prints the value of the attributes.

E.g.
```
blue_circle = Shape('circle', 'blue')
green_square = Shape('square', 'green')

blue_circle.get_info()
green_square.get_info()

```
    
prints:

    This is a circle and it is blue
    This is a square and it is green

````python
# Solution for exercise 1

class Shape:

    def __init__(self, name, colour):
        self.name = name
        self.colour = colour
        
    def get_info(self):
        print("This is a", self.name, "and it is", self.colour)

blue_circle = Shape('circle', 'blue')
green_square = Shape('square', 'green')

blue_circle.get_info()
green_square.get_info()
````

## Exercise 6.2
Create a class `Circle` that inherits from `Shape` and has the additional attribute `radius` as well as a method `get_area(self)` that prints the area of the circle.

E.g.:
```
blue_circle = Circle('blue', 1)
red_circle = Circle('red', 2)

blue_circle.get_info()
blue_circle.get_area()
red_circle.get_info()
red_circle.get_area()
```
    
prints:

    This is a circle and it is blue
    The area of the blue circle is 3.141592653589793
    This is a circle and it is red
    The area of the red circle is 12.566370614359172

````python
# Solution for exercise 2

from math import pi

class Shape:

    def __init__(self, name, colour):
        self.name = name
        self.colour = colour
        
    def get_info(self):
        print("This is a", self.name, "and it is", self.colour)

class Circle(Shape):

    def __init__(self, colour, radius):
        Shape.__init__(self,'circle', colour)
        self.radius = radius
        
    def get_area(self):
        area = pi * self.radius**2
        print("The area of the", self.colour, "circle is", area)

        
        
        
blue_circle = Circle('blue', 1)
red_circle = Circle('red', 2)

blue_circle.get_info()
blue_circle.get_area()
red_circle.get_info()
red_circle.get_area()
````

## Exercise 6.3
Create a method `getInfoFromAll(self)` in the class `Shape` that calls the method `get_info()` for each shape object created so far and, if it is a circle, it also calls `get_area()`.

E.g.:

```
blue_circle = Circle('blue', 1)
green_square = Shape('square', 'green')
yellow_triangle = Shape('triangle', 'yellow')

blue_circle.get_info_from_all()
```
    
prints:

    This is a circle and it is blue
    The area of the blue circle is 3.141592653589793
    This is a square and it is green
    This is a triangle and it is yellow

````python
# Solution for exercise 3

from math import pi

class Shape:
    
    shapes = []

    def __init__(self, name, colour):
        self.name = name
        self.colour = colour
        Shape.shapes.append(self)

    def get_info(self):
        print("This is a", self.name, "and it is", self.colour)
        
    def get_info_from_all():
        for s in Shape.shapes:
            s.get_info()
            if type(s) == Circle:
                s.get_area()

class Circle(Shape):

    def __init__(self, colour, radius):
        Shape.__init__(self, 'circle', colour)
        self.radius = radius

    def get_area(self):
        area = pi * self.radius**2
        print("The area of the", self.colour, "circle is", area)

blue_circle = Circle('blue', 1)
green_square = Shape('square', 'green')
yellow_triangle = Shape('triangle', 'yellow')

Shape.get_info_from_all()
````

## Exercise 6.5
There are two tutors, and each tutor teaches three students. Define a `Tutor` class which stores his/her students and has a function to compute the mean of his/her students’ scores (you can use the class in exercise 4.5). 


````python
class Tutor:
    def __init__(self):
        self.students=[]
    
    def add_student(self, student):
        self.students.append(student)
    
    def mean(self):
        total=0
        for s in self.students:
            print(vars(s))
            total += s.score
        
        print(total/len(self.students))

tutor1 = Tutor()
tutor2 = Tutor()

for s in students[0:3]:
    tutor1.add_student(s)

for s in students[3:6]:
    tutor2.add_student(s)

tutor1.mean()
tutor2.mean()
````

## Exercise 6.6
Create a bank customer `Account` class which can handle transfer of money to a receiver, so that the class can allow to specify the amount of money to be transferred, and display the new balances of sender and receiver. 


````python
class Account:
    def __init__(self, name, start_funds):
        self.name = name
        self.balance = start_funds
    
    def transfer_to(self, receiver, amount):
        
        if type(receiver) != Account:
            raise TypeError("You cannot transfer money to something that is not an Account object.")
        
        if amount <= 0:
            raise ValueError("Amount to transfer must be positive.")
        
        if amount <= self.balance:
            receiver.balance += amount
            self.balance -= amount
            print("Transfer successful: {0} --( £{2:,.2f} )--> {1}".format(self.name, receiver.name, amount))
        else:
            print("Transfer failed. Insufficient funds.")
            
    def display_balance(self):
        print("Current balance of {0}: £{1:,.2f}".format(self.name, self.balance))
    
dupont = Account("Dupont", 1000)
smith = Account("Smith", 100)

smith.transfer_to(dupont, 100)

dupont.display_balance()
smith.display_balance()
````

## Exercise 6.7
Create a `Bank` class which stores the above `Account` objects in an internal database.


````python
class Bank:
    def __init__(self, bank_name, customers):
        self.name = bank_name
        self.accounts = []
        for c in customers:
            self.accounts.append(Account(c[0], c[1]))
    
    def total_balance_stored(self):
        total_stored = 0
        for customer in self.accounts:
            total_stored += customer.balance
        return total_stored
    
    def display_total_balance(self):
        print("The total balance stored by {name} is £{balance:,.2f} which is from {n_customers} customers."
             .format(name=self.name, balance=self.total_balance_stored(), n_customers=len(self.accounts)))
    
nateast = Bank("Nateast", [("Smith", 1000000), ("Dupont", 20000)])
nateast.display_total_balance()
````

## Exercise 6.8
Write a `Company` class which pays a salary to their employees' accounts.


````python

class Company:
    def __init__(self, name):
        self.name = name
        self.account = Account(name, 10000000) # We need to create a company account
        self.employees = [] # our employee accounts will be stored here when they are hired
        self.salary_rate = 30000 # this is the rate we will pay all our employees at
        
    def add_employee_account(self, e_account):
        if e_account not in self.employees:
            self.employees.append(e_account)
            print("Employee account {} has just been added to payroll.".format(e_account.name))
        else:
            print("This employee account has already been added to payroll.")
                
    def pay_salaries(self):
        for employee in self.employees:
            self.account.transfer_to(employee, self.salary_rate)
        

smith = Account("Smith", 10000)
smith.display_balance()

company = Company("Macrohard")
company.add_employee_account(smith)
company.add_employee_account(smith)
company.pay_salaries()

smith.display_balance()
````
