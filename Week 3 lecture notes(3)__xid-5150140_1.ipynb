Let's now have a look at the difference between lists and arrays for a moment.

Imagine, I want to pass a list to a function to obtain a list of outputs. The math library cannot support this without using a loop.

```python
#This won't work
from math import sin 
def vecsin(L):
    y = sin(L)
    return y

d = [1,2,3,4,6,8,10]
out = vecsin(d)
print(out)
```

I have to pass elements of the list to the function one by one in a loop.

```python
#This will work
from math import sin 
def vecsin(L):
    y = sin(L)
    return y

d = [1,2,3,4,6,8,10]
out = []
for i in range(len(d)):
    out.append(vecsin(d[i]))

print(out)
```

But with numpy sin function, we can pass a list to a function as an argument to obtain a list output.

```python
import numpy as np
def vecsin(L):
    y = np.sin(L)
    return y

d = [1,2,3,4,6,8,10]
out = vecsin(d)
print(out)
```

It is easy to perform addition, multiplication, or division on Python arrays.

```python
import numpy as np
x = np.arange(1,5)
y = np.arange(2,6)
print("Length of x is ",len(x),", and the legth of y is ",len(y))
add = x + y
prod = x*y
div = x/y

print('Addition of {} and {} = {}' .format(x,y,add))
print('Product of {} and {} = {}' .format(x,y,prod))#Element wise multiplication of two vectors
print('Division of {} and {} = {}' .format(x,y,div))#Element wise division of two vectors
```

```python
#Finding the maximum and minimum in an array

import numpy as np #Import numpy as np to be short
#Then use the random number generating library of functions like rand, randint
intlist = np.random.randint(10,size=(8,3))
print('Integer list is: ',intlist)
#Then find the maximum value in the array
maxint = np.max(intlist)
print('Maximum value of the list is: ',maxint)
index = np.argmax(intlist)
print('Elemnt ',index+1,' had the first max value')
whichn = np.where(maxint == intlist) #This gives the indexes of where it is located in the array 
#It is useful to locate multiple occurances
print('Indexes of all the max values? ',whichn)

#Find the minimum
minint = np.min(intlist)
print('The minimum value of the list is: ',maxint)
index = np.argmin(intlist)
print('Element ',index,' had the first min value')
whichn = np.where(minint == intlist) #This is another way to find where it is in the array
print('Indexes of all the min values? ',whichn)
```

## 3.2 Using the linear algebra module

```python
import numpy as np
a = np.random.randint(5,size=(3,3))
ina = np.linalg.inv(a) #Use inverse from linear algebra module
deter = np.linalg.det(a) #Use determeinat from linear algebra module
R = np.linalg.matrix_rank(a) #Use rank from linear algebra module
[E,V] = np.linalg.eig(a)
print('Matrix a: {}' .format(a))
print('The inverse of a: {}' .format(ina))
print('The determinant of a: {}' .format(deter))
print('The rank of a: {}' .format(R))
print('The Eigen values of a: {}' .format(E))
print('The Eigen vectors of a: {}' .format(V))
```

```python
a = np.random.rand(3,4)
pina = np.linalg.pinv(a) #Use pseudo-inverse from linear algebra module
print('Matrix a: {}' .format(a))
print('The pseudoinverse of a: {}' .format(pina))
```

## 3.6 Using time

```python
import time
import numpy as np
tt = []
for i in np.arange(1,10):
    t0 = time.time()
    time.sleep(1) #Sleep 1 second
    t = time.time()
    #y = time.ctime() #Clock time
    print("The elapsed time is ", (t-t0), "[sec]")
    tt.append(t-t0)
avt = np.average(tt)
print("Average delay is {:.4f} [sec]" .format(avt))
```

## 3.7 Symbolic computation in python
We can use Python coding to do operations like differentiation and integration where we often end up making mistakes if we do it manually. Take for instance differentiating an equation like the vertical position of a ball moving under gravity given by $y(t) = v_0 t - \frac{1}{2}gt^2$ to obtain it's velocity.

* $y$ is the vertical position as a function of time $t$
* $v_0$ is the initial vertical velocity (at $t=0$)
* $g$ is the acceleration due to gravity

```python
#We can try to substitute values to the projectile equation mentioned above
from sympy import *
#We first define the symbols
v0,t,g,y,v = symbols('v0 t g y v')

y = (v0*t - 0.5*g*t**2) #This is now a symbolic equation
print(y)
#Then we can use substitution to find the value of the function 
yt = y.subs([(t,2),(v0,10),(g,9.81)])

print('yt: {:.2f} m'.format(yt))
```

### Learn how to solve an equation, i.e. $y = f(x)$, find $x$ when $f(x) = 0$.

```python
#Here we first import solve function from solver package 
#in sympy module
from sympy.solvers import solve 
#Then from sympy we import the submodule symbol
#that allows us to define symbolic variables
from sympy import symbols
#We first define the symbols
v0,t,g,y,v = symbols('v0 t g y v')
y = (v0*t - 0.5*g*t**2);
roots = solve(y, t) #solve(function,t) solves the function with respect to t
#Solving is to find t at which y = 0
#The solutions will give us the times at which the object will be on the ground level
print('Roots of y = 0 are t = {} and t = {}' .format(roots[0],roots[1]))
r2 = roots[1].subs([(v0,2),(g,9.81)])
print('t2: ',r2)
```

### Differentiation of an equation

```python
from sympy import *
#We can perform other computations like differentiation
v0,t,g,y,v = symbols('v0 t g y v')
y = v0*t - 0.5*g*t**2 #Position
vel = diff(y,t,1)#1 tells the differentiation algorithm in the library to differentiate only once.
#Try to differentiate twice to get the acceleration.
print(vel)
vt2 = vel.subs([(t,2),(g,9.81)])
print('Velocity at t = 2 is {}' .format(vt2))
vt2v0 = vel.subs([(t,2),(g,9.81),(v0,10)])
print('Velocity at t = 2 for v0 = 10 is {:.2f}' .format(vt2v0))
```

```python
from sympy import *
v0,t,g,y,v = symbols('v0 t g y v')
y = v0*t - 0.5*g*t**2 #Position
vel = diff(y,t,1)#1 tells the differentiation algorithm in the library to differentiate only once.
#Try to differentiate twice to get the acceleration.
print('Velocity equation: ',vel)
acc = diff(y,t,2)
print('Acceleration equation: ',acc)
vt = v0 + integrate(acc,t) #We have to add initial velocity
print('Velocity after integrating acceleration wrt time: ',vt)
yt = 0 + integrate(vt,t) #We have to add initial position
print('Position after integrating velocity wrt time: ',yt)
```
