```python
import matplotlib.pyplot as plt 
from math import *
import numpy as np

def Coordinates(theta):
    x = sin(theta)
    y = cos(theta)
    return x,y

f = np.vectorize(Coordinates) #Vectorize the function - Coordinates - to obtain array outputs given 
#array inputs, and assign it to a new function name called f. Note, it can be any name.

theta = np.arange(0,2*pi,pi/100)
x,y = f(theta) #Pass theta to function f and obtain the coordinates x and y as arrays.
plt.plot(x,y,'o:')
plt.show()
```

```python
from math import *
import numpy as np
import random as r

ch = np.array(['A','B','C','1','2','3'])
print(ch)

def CodeGen(ch):
    N = len(ch)
    code = []
    for i in range(N):
        number = r.randint(0,N-1) #Use N-1 because indexes start from 0
        code.append(ch[number])
    return code #Note the indentation is right under the "for" to make sure returning is not part of the for loop

for j in range(10):
    code = CodeGen(ch)
    print(code)
    


```

```python
from sympy import *
import matplotlib.pyplot as plt 
from math import *
import numpy as np

x = Symbol('x') #define x as a symbolic variable

def MyFun(x):
    f = x**3 - 12*x**2 + 44*x -48
    return f

f = MyFun(x)
fd = diff(f,x,1) #Take the first differential of f with respect to x

VecMyFun = np.vectorize(MyFun) #Vectorize MyFun and assign it to VecMyFun that can
#take vector inputs and give out vector outputs

x = np.arange(1,6,0.1) #generate values for x between -2 and 6 in steps of 0.1
fx = VecMyFun(x)

sol = solve(fd)
s1 = sol[0].evalf()
s2 = sol[1].evalf()

print("Solutions are %1.2f and %1.2f" %(s1,s2))
plt.plot(x,fx)
plt.show()

#Observation of the function f makes it clear that the function has a local maximum and a minimum at the solutions for
#the differentiated function.
```
