# 2. Functions
Programmers realised that when writing code, there are a lot of occasions where you want to do the same computation again and again - but it’s quite annoying and messy to rewrite all your code each time you would like to use it. The concept of a function solves this issue! A function is something which you can “call” in your code and it will perform a specified function! Hence the name!

In Python, the syntax to use a function looks like this:

**FunctionName(arguments)**

'arguments' are inputs you pass to the function to receive an output. 

It’s pretty simple - you just “call” the function by its name and you give the correct inputs in the right order in parenthesis next to it. If you’re confused about what arguments a function requires, then you can always look it up or ask.

**output = FunctionName(arguments)**

Let's look at few examples.

```python
#A function gets input values and passes outputs
#We define a function like below
#
#'myfun' is the name of the function
def myfun(name):
    print('Came inside the function')
    text = 'Hello ' + name + '!' 
    #remember to use the : at the end of the function
    #In this case, 'name' is the argument or the input to the function
    return text #This is what the function returns

#When you run the code, it will start from function calls not the function itself.
print('Call myfun for Anne')
output = myfun('Anne') #Then we can call the function by passing a text string for the name
print(output)
print('Call myfun for Thrish')
output = myfun('Thrish') #Then we can call the function by passing a text string for the name
print(output)
print('Call myfun for Nick')
output = myfun('Nick') #Then we can call the function by passing a text string for the name
print(output)
```

**Note**: Check the order of the above displays. First, the statement "Call myfun for Anne" is displayed. Then the print statement inside the function "Came inside the function" is displayed.

Then the function returns the result, which is assigned to the output of the function.

Note that the code segment for the function is not executed till it is called somewhere in the program.

```python
#A function gets input values and passes outputs
#We can pass more than one value
def TakePower(value,exp):
    pw = pow(value,exp)
    return pw #This will return the product back to the caller.

a = 3
b = 2
op = TakePower(a,b) #This is how you call the multi function.
print("The power of ",a," to ",b, " is: ", op)
```

Now let us try to use the idea of a function to evaluate $y(t) = v_0t- \frac{1}{2}gt^2$ for $v_0=5$, $g=9.81$ and $t=0.6$. If you were doing this on paper you would probably write something like this: $$ y = 5\cdot 0.6 - {1\over2}\cdot 9.81 \cdot 0.6^2.$$ Happily, writing this in Python is very similar:

```python
#First we define the function called projectile
def projectile(v0,t):
    print('Memory location of t is: ',id(t))
    print('Memory location of v0 is: ',id(v0))
    g = 9.81
    y = v0*t - 0.5*g*t**2
    print('Memory location of y is: ',id(y))
    return y,g

#Then we assign values to input variables also known as arguments
#AccelGravity = 9.81
time = 0.6
print('Memory location of time is: ',id(time))
InitialVel = 5
print('Memory location of InitialVel is: ',id(InitialVel))
#Then we call the function and pass the arguments.
#The output of the function is assigned to the position variable
position,AccelGravity = projectile(InitialVel,time) #Passes y to position, and g to AccelGravity
print ("The position at time {} sec is: {:.2f} m" .format(time,position))
print ('If I throw a stone on earth '+
       '(average acceleration of gravity is {:.2f} m/sec^2), at initial velocity ' .format(AccelGravity)+
       '{:.1f} m/sec, it will be at height {:.2f} m at time {} sec' .format(InitialVel,position,time) + 
       4*'!')
```

## 2.1 Changing the value of a variable inside a function without affecting the original value

**Note**: Arguments have two components. 1) The address of the memory location, 2) The value

In Python, when we pass an argument to a function, we pass copies of the address and the value. It does not erase the original value at the address. In otherwords, passing a variable as an argument of a function is **not a destructive** operation.

```python
from IPython.display import Image
Image(filename='PassingByReference.jpg')
```

```python
def f(s1): #The function f takes s as an argument
    print('Address of s1 before the new assignment is: ', id(s1)) #Address of s
    s1 = "I love cats!" #The function then modifies s
    print(s1,' inside the function') 
    print('Address of s1 after the new assignment is: ', id(s1))

s = "I love dogs!" #A string value assign to s
print(s,' before calling the function') #Print the value of s before passing
print('Address of s before calling the function: ', id(s))
f(s) #Call function f by passing s
print(s,' after calling the function') #Print the value of s after passing
print('Address of s after calling the function: ', id(s))
```

Please note that the function changed the string value assigned to the variable s inside the function, but it did not affect the original string assigned to s before calling the function.

What happens here, is that the variable name s points at the string value "I love dogs" first. Then when s is passed to the function f, the string value assigned to s is passed as the **argument** (input) of the function. Any change to that value inside the function causes a change of address. Thus it does not affect the original value pointed by s. 

An analogy is shown in the above figure, where a copy of the circular value in a container (in our case, the container is the variable s, and the circular value is the string "I love dogs") is passed to the function. The function then converts it to a square. But that does not change the original value. The new shape is valid only inside the function though it is refered to have come from the same container.

```python
#Take another example
def reassign( d1 ):
    "This changes a passed list into this function"
    print("Values before assignment inside the function: ", d1)
    print('Address of d1 before reassignment inside the function is: ',id(d1))
    d1 = [1,2] #Assign a new list to d1
    print("Values inside the function: ", d1)
    print('Address of d1 after reassignment inside the function is: ',id(d1))
    return d1

# Now you can call changeme function
d = [10,20,30];
print("Values before calling the function: ", d)
print('Address of d before calling the function is: ',id(d))
d2 = reassign( d );
print("Output of the function is: ", d2)
print('Address of d2 is: ',id(d2))

print("Values after calling the function: ", d)
print('Address of d after calling the function is: ',id(d))
```

## 2.2 Changing the value of a variable at the same memory location
In Python, lists that are modified inside a function affect the original list too. This is because, a list, unless replaced entirely, will keep it's original memory location in the CPU. An analogy is keeping the container in the above example.  

```python
def reassign( v1 ):
    "This changes a passed list into this function"
    print("Value of v1 before assignment inside the function: ", v1)
    print('Address of v1 before modification is: ',id(v1))
    v1.append(40)
    print("Value of v1 inside the function: ", v1)
    print('Address of v1 before modification is: ',id(v1))
    #return

# Now you can call changeme function
v = [10,20,30];
print("Value of v before calling the function: ", v)
print('Address of v before calling the function is: ',id(v))
reassign( v );
print('Address of v after modification is: ',id(v))
print("Value of v outside the function: ", v)
```

```python
def reassign( v1 ):
    v2 = v1.copy()
    "This changes a passed list into this function"
    print("Value of v1 before assignment inside the function: ", v1)
    print('Address of v1 before modification is: ',id(v1))
    
    print("Value of v2 before assignment inside the function: ", v2)
    print('Address of v2 before modification is: ',id(v2))
    
    v2.append([1,2])
    print("Value of v1 inside the function: ", v1)
    print('Address of v1 before modification is: ',id(v1))
    
    print("Value of v2 inside the function: ", v2)
    print('Address of v2 before modification is: ',id(v2))
    #return

# Now you can call changeme function
v = [10,20,30];
print("Value of v before calling the function: ", v)
print('Address of v before calling the function is: ',id(v))
reassign( v );
print('Address of v after modification is: ',id(v))
print("Value of v outside the function: ", v)
```

In the above example, we can see that the address of v did not change when it was copied to v1 in the function. This is because both the address and the value are passed to the function. Then, when the function appended values to the list, it overwrote the new list in the same memory location because lists allow this to happen. 

```python
def f(s1): #The function f takes s as an argument
    print('Address of s before the new assignment is: ', id(s1)) #Address of s
    s1 = 'I love cats!' + ' for a while' #The function then modifies s
    print(s1,' inside the function') 
    print('Address of s after the new assignment is: ', id(s1))

s = "I love dogs!" #A string value assign to s
print(s,' before calling the function') #Print the value of s before passing
print('Address of s before calling the function: ', id(s))
f(s) #Call function f by passing s
print(s,' after calling the function') #Print the value of s after passing
print('Address of s after calling the function: ', id(s))
```

We notice that even an addition of text on top of the original text will cause the new content to be placed in a new memory location without affecting the original content.

In Summary, you can safely remember that only modification of lists inside a function happens at the same memory location in Python.

# 3. Python modules

The people who created Python realised that there are many functions which people use very regularly. Therefore, they wrote the code for all of those functions and put them into modules. A module is a collection of related functions, variables, and classes (we will study what classes are later) packaged to make your life easier. 

You can access all functions written in a **module** using the “import” statement. A good analogy for this is to imagine boxes with labels.

A module is a file containing Python definitions and statements. You should import a module when you need to use any of the functions that are defined inside it (e.g., if you need to perform a square root, you should import the module 'math' and use the function 'sqrt' that is defined inside it).

In Python, the terms 'library' and 'module' are interchangable most of the time. You will often hear people say 'library' instead of 'module' because it is a more widespread term and is used in other programming languages.

The 'import' statements are usually written in the first lines of code. There are several ways to import a module in Python (all of them are valid), listed below:

# 4. Conditions

Very often we want to execute instructions when some condition is true, like proceed to check the bank balance if the pin code matches the account details. Here, we learn how to use IF statements FOR and WHILE to control such conditions.

```python
x = True
y = False

if x: #This means, if x is True, do the following:
    print('The condition is ',x,'. execute this first print command')
elif y: #Else if y is True, do the following:
    print('The condition is ',y,'. execute this second print command instead') 
```

```python
x = False
y = True

if x:
    print('Execute this first print command')
elif y:
    print('Execute this second print command instead') 
```

```python
#Check what happens when both conditions are True
x = True
y = True

if x:
    print('Execute this first print command')
elif y:
    print('Execute this second print command instead') 
```

```python
#Check what happens when both conditions are False
x = False
y = False

if x:
    print('Execute this first print command')
elif y:
    print('Execute this second print command instead') 
```

```python
#The boolean conditions of True or False can be created by comparison of numbers too
#Check the following scenario
x = int(input('Enter a number: ')) #The entered values will be converted to an integer and assigned to x
y = int(input('Enter another number: ')) #Similarly

if x > y: #if CONDITION: The block of instructions under if CONDITION will be executed if CONDITION is TRUE
    print('{} is greater than {} is {}' .format(x,y,x>y))
    print('{} is less than {} is {}' .format(x,y,x<y))
elif x < y: #if CONDITION is FALSE, the code will skip the above two print commands and check this ELSE IF (elif) condition
    print('{} is greater than {} is {}' .format(x,y,x>y)) #These two print commands will be executed if the second condition is true
    print('{} is less than {} is {}' .format(x,y,x<y))
```

```python
#You can use what you learnt in functions to repeat asking the question
def CheckNum(x,y): #Note the colon and the indentation of the rest of the instructions

    if x > y: #Note the colon and the indentation of the rest of the instructions
        print('{} is greater than {} is {}' .format(x,y,x>y))
        print('{} is less than {} is {}' .format(x,y,x<y))
    elif x < y: #Note the colon and the indentation of the rest of the instructions
        print('{} is greater than {} is {}' .format(x,y,x>y))
        print('{} is less than {} is {}' .format(x,y,x<y))
        
    cond = input('Want to play again? Enter y or n: ')
    if cond == 'y': #Note the colon and the indentation of the rest of the instructions
        x = int(input('Enter a number: '))
        y = int(input('Enter another number: '))
        CheckNum(x,y) #The same function can be called from within the function
    elif cond == 'n':
        return
    else: #Note the colon and the indentation of the rest of the instructions
        return 

x = int(input('Enter a number: '))
y = int(input('Enter another number: '))
#We call the function once
CheckNum(x,y)
```
