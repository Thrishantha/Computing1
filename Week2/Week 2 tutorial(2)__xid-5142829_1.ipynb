# Inverted pendulum balancing project

There are many real life control problems like biped walking with an upright trunk, the Segway, or path tracking of a rocket, that can be approximated by an inverted pendulum. In an inverted pendulum, a pendulum mounted on a cart with a free pivot joint should be balanced upright by pushing or pulling the cart. In this case, we use the principle of inertia. The moment of inertia of a point mass further away from the pivot joint is larger than that of a point mass closer to the pivot joint. So, when the cart is pushed in one direction, the point mass closer to the pivot joint will move more than the point mass further away. Since the point masses are located along a rigid rod, the pendulum will rotate in the opposite direction of the cart's movement. This principle can be used to balance the pendulum by moving the cart.  

We will develop a huristic rule based controller for the inverted pendulum when we progress through future tutorials. In this tutorial, we take the first step of coding the dynamic equations of the inverted pendulum.

## Inverted pendulum control, step-1: 
### Dynamic equations of the inverted pendulum
Let cart position from the centre point of the rail be $x$ [m], it's speed be $\dot{x}$ [m/sec], and it's acceleration be $\ddot{x}$ [m/sec$^2$].

Let pendulum angle from the upright balanced posture be $\theta$ [rad], it's speed be $\dot{\theta}$ [rad/sec], and it's acceleration be $\ddot{\theta}$ [rad/sec$^2$].

Let the mass of the pendulum be $m$ [kg], the length of the pendulum be $l$ [m], the moment of inertia be $I$ [kg m $^2$], the mass of the cart be $M$ [kg], and the coefficient of friction of the cart on the rail be $\mu$.

Let the external force given to push or pull the cart be $F$ [N].

Then, the dynamic equations of an inverted pendulum are given by:
$$\ddot{x} = -\frac{(I + m l^2) \mu}{d} \dot{x} + \frac{(m l)^2 g}{d} \theta + \frac{(I+m l^2)}{d} F$$

where, $g = 9.81$ [m/sec$^2$] is the acceleration of gravity, $d = I (M+m) + M m l^2$.

$$ \ddot{\theta} = -\frac{m l \mu}{d} \dot{x} + \frac{(M+m) m g l}{d} \theta + \frac{m l}{d} F$$

Given $m = 0.1$ [kg], $M = 1$ [kg], $l = 1$ [m], $I = 0.5 m l^2$ [kg m$^2$], $\mu = 0.3$, and $g = 9.81$ [m/sec$^2$], write a function that returns $\ddot{x}$ and $\ddot{theta}$ given above parameters and $\dot{x}$, $\theta$, and $F$.
